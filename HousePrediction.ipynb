```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from mpl_toolkits.mplot3d import Axes3D
class HousePrediction(object):


    # Sample data (you can replace this with your dataset)
    data = {
        'SquareFeet': [1500, 1600, 1700, 1800, 1900, 2000, 2100, 2200],
        'Bedrooms': [3, 3, 3, 4, 4, 4, 5, 5],
        'Price': [300000, 320000, 340000, 360000, 380000, 400000, 420000, 440000]
    }

    df = pd.DataFrame(data)

    # Features and Target
    X = df[['SquareFeet', 'Bedrooms']]
    y = df['Price']

    # Fit the model
    model = LinearRegression()
    model.fit(X, y)

    # Create grid for plotting surface
    x_surf, y_surf = np.meshgrid(
        np.linspace(df['SquareFeet'].min(), df['SquareFeet'].max(), 10),
        np.linspace(df['Bedrooms'].min(), df['Bedrooms'].max(), 10)
    )
    z_surf = model.predict(np.c_[x_surf.ravel(), y_surf.ravel()]).reshape(x_surf.shape)

    # 3D Plot
    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection='3d')

    # Plot the surface
    ax.plot_surface(x_surf, y_surf, z_surf, cmap='viridis', alpha=0.7)

    # Plot actual data points
    ax.scatter(df['SquareFeet'], df['Bedrooms'], y, color='red', label='Actual Prices')

    ax.set_xlabel('Square Feet')
    ax.set_ylabel('Bedrooms')
    ax.set_zlabel('Price')
    ax.set_title('House Price Prediction Surface')

    plt.legend()
    plt.show()


```
